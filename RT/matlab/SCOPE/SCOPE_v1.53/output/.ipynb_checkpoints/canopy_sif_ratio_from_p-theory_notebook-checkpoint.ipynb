We are attempting to simulate the total escaping sun-induced fluorescence (SIF) from the canopy, in effect we try to imitate the canopy fluorescence radiative transfer. The leaf level SIF is obtained together with the leaf single scattering albedo from the leaf biochemical and leaf radiative transfer module within SCOPE. The parameters required are the fluorescence matrices and efficiency terms for both PSI and PSII. At this stage we simply use average values of the efficiency for both in sun and shade leaves as well as averaging over the depth of the canopy. PSI does not require a factor for efficiency as it is known to be a reasonably unaffected by NPQ compared to PSII and PSI has a reasonably constant fluorescence yield (Porcar-Castell et al, 2014). Matrices are simply integrated over the excitation spectrum to yield a leaf level SIF curve per fluorescence wavelength. No separation into fractions of sunlit and shaded has been attempted at this stage. We see this as a proof of concept of the use of p-theory to describe the photon escape path from the canopy and as such further research will be required to introduce the use of this separation. 


The following code came about from an attempt to model the SIF signal from leaf to canopy level. The SIF signals are not in units of $Wm^{-2}$ but in a unitless ratio proportional to the integral of the intercepted incomming radiation between 640 nm and 850 nm. The SCOPE code has been forked to extract the relevant datasets required for our model. The SCOPE code was run with varying values for canopy structural parameters. The output folders are saved in the same directory as this notebook. Each folder name starts with the value of the structural parameter used for the run, eg LAI_6_erect_2014-11-07-1242/ for a LAI value of 6, and an erectophile leaf angle distribution, and the date and time of the run. Additional python and bash scripts in the same folder are required for formating and also some calculations. The scripts are as follows with a description:
1. mac2unix_all.sh              
SCOPE saves all *.dat files in mac format which needs to be converted to unix format.
2. convert_mW_W.sh    
Simply converts a file from units of $mWm^{-2}$ to $Wm^{-2}$.
3. canopy_interceptance.py     
Calculates the canopy interceptance ($i_0$) based on Stenberg (2007). Interceptance is used to correct for incoming irradiance as well as to calculate the canopy recollision propability ($p$).
4. canopy_albedo.sh    
Calculates the canopy level albedo based on outgoing radiation and incoming intercepted irradiance. 
5. canopy_sif_ratio2.sh   
Calculates the canopy level SIF ratio based on outgoing SIF radiation which is derived from the SAIL radiative transfer model. This is the quantity we are trying to compare against using our p-theory method. 
6. canopy_sif_model.py   
This is the model which we describe below.

There is no need to run these scripts again unless new SCOPE runs with different structural parameter values are attempted creating new folders. Note that all outgoing quantities such as canopy albedo and SIF ratio are spherical (not hemispherical) as both upward and downward radiation is included. This differs from the usual standard which considers only the upper-hemispherical quantities. Soil reflectance has been set to zero for all simulations in SCOPE.

A spectral invariant approach to estimating canopy level SIF ratio as portion
of intercepted radiation. This model requires the p_theory.py module to be in
the pythonpath. It requires as a single argument, the path to the following 
generated files for each parameter in the model:   
* canopy_spectral_albedo.dat              
Wc
* leaf_spectral_albedo.dat                
wl
* leaf_spectral_fluorescence_mat.dat      
fsr
* canopy_spectral_sif_irr_ratio.dat       
SIFr
* wl.dat                                  
lam
* Parameters/inputdata.txt                
LAI, LIDFa, LIDFb

```python
# read parameters from local files
Wc = np.genfromtxt(os.path.join(root, 'canopy_spectral_albedo.dat'))
wl = np.genfromtxt(os.path.join(root, 'leaf_spectral_albedo.dat'))
fsr = np.genfromtxt(os.path.join(root, 'leaf_spectral_fluorescence_mat.dat'))
SIFr = np.genfromtxt(os.path.join(root, 'canopy_spectral_sif_irr_ratio2.dat'))
lam = np.genfromtxt(os.path.join(root, 'wl.dat'),skip_header=2)

# reads in parameters from the inputdata.txt file
filename = os.path.join(root,'Parameters/inputdata.txt')
f = open(filename, 'r')
lines = f.read()
f.close()
i = lines.index('LAI')
LAI = np.float(lines[i+3:i+7])
i = lines.index('LIDFa')
LIDFa = np.float(lines[i+5:i+11])
i = lines.index('LIDFb')
LIDFb = np.float(lines[i+5:i+11])
```

Below is the calculation of the recollision probability ($p$). This is based on Stenberg (2007) paper. This is a simple analytical formula for calculating $p$ based on $i_0$ and LAI. It is only valid for a turbid medium approach.
$$ p = 1 - \frac{i_0}{LAI} $$
Where $i_0$ is based on the canopy gap probability.
$$ i_0 = \frac{1}{2\pi} \int_0^{2\pi}\mathrm{1 - P_0(\Omega)}\mathrm{d}\Omega$$
$P_0$ is gap probability based on Myneni et al (1989). We have implimented the code to include leaf spatial distributions, but as the SCOPE code does not a random Poisson distribution is assumed. Our method uses a fast and accurate Gaussian quadrature approach for the angular intergration.

```python
# calculate p recollision probability
p = pt.p(LAI, arch='l', par=(LIDFa, LIDFb))
```

Following is a spectral invariant estimate of the leaf canopy albedo following Lewis & Disney, (2007) Panferov et al, (2001) Knyazikhin et al, (2011) Smolander & Stenberg, (2003) Mottus, (2010) and others.
$$ W = \frac{\omega(1 - p)}{1 - p \omega} $$
Where $\omega$ is the leaf single scattering albedo.

```python
# calculate the spectral invariant estimated canopy albedo
We = pt.Alb(wl, p)
```

```python
# plot comparison between estimated canopy and SCOPE canopy albedo
beg = 0
end = 451
plt.title(r'Canopy and Leaf Spectral Albedo', fontsize=16)
plt.plot(lam[beg:end], wl[beg:end], label='PROSPECT Leaf Albedo')
plt.plot(lam[beg:end], We[beg:end], label='p-theory Canopy Albedo')
plt.plot(lam[beg:end], Wc[beg:end], label='SCOPE Canopy Albedo')
plt.xlabel(r'Wavelength in nm ($\lambda$)', fontsize=14)
plt.ylabel('Albedo', fontsize=14)
plt.legend(loc=2)
plt.show()
```

```python
# plot comparison between estimated canopy and SCOPE canopy albedo excluding leaf surface scattering
beg = 0
end = 451
plt.title(r'Canopy and Leaf Spectral Albedo', fontsize=16)
winf = np.min(wl[beg:end])
wl_ = (wl - winf) / (1 - winf)
plt.plot(lam[beg:end], wl_[beg:end], label='PROSPECT Leaf Albedo')
plt.plot(lam[beg:end], We[beg:end], label='p-theory Canopy Albedo')
plt.plot(lam[beg:end], Wc[beg:end], label='SCOPE Canopy Albedo')
plt.xlabel(r'Wavelength in nm ($\lambda$)', fontsize=14)
plt.ylabel('Albedo', fontsize=14)
plt.legend(loc=2)
plt.show()
```

```python
plt.scatter(wl[beg:end],Wc[beg:end], c='r', label='SCOPE canopy albedo')
plt.scatter(wl[beg:end],We[beg:end], c='g', label='p-theory canopy albedo')
plt.xlim(0,1)
plt.ylim(0,1)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('Leaf single scattering albedo ($\omega$)', fontsize=14)
plt.ylabel('Canopy albedo ($W$)', fontsize=14)
plt.title('Leaf albedo vs canopy albedo', fontsize=16)
plt.legend(loc=2)
plt.show()
```

The following is the spectral invariant approach to estimating the canopy SIF ratio. We have found that the 4th order approximation gives the best results. The Newmann equation exact solution seems to overestimate SIF in the high scattering near infra red (NIR) region, thus over the second fluorescence peak at 730 nm. To use the exact solution uncomment and comment the lines below. The exact solution is:
$$ E = \frac{\omega(1 - p)}{1 - p \omega} - (1 - p)\Big(\frac{\omega + \epsilon - 2\epsilon \omega p - \omega ^2 p + 2 \epsilon p^2 \omega ^2 - \epsilon ^2 p + 2 \epsilon ^2 p^2 \omega}{(1 - \epsilon p)^2 (1 - \omega p)^2}\Big) $$
The solution to the 4th order is:
$$ E = (1 - p)\Big(\epsilon + \epsilon ^2 p + 2 \epsilon p \omega + \epsilon ^3 p^2 + 3 \epsilon ^2 p^2 \omega + 3 \epsilon p ^2 \omega ^2 + 4 \epsilon ^3 p^3 \omega + 6 \epsilon ^2 p^2 \omega ^2 + 4 \epsilon p^3 \omega ^3 + \epsilon ^4 p^3 \Big) $$
E is the canopy level SIF ratio, and $\epsilon$ is the leaf level SIF ratio.
The solution was derived from the successive order of scattering approach whereby each stream is separated into a scattering or absorption stream. Below is a graph of the successive orders of scattering within the canopy.

<img src="files/SOSA_SIF.png">

```python
# calculate the spectral invariant estimate of canopy SIF ratio
beg = 240
end = 451
#SIFre = pt.SIFr(wl[beg:end], p, fsr)
SIFre = pt.SIFr_approx(wl[beg:end], p, fsr) 
```

```python
# plot comparison between estimated canopy and SCOPE canopy SIF ratio
beg = 240
end = 451
plt.title('Canopy and Leaf Spectral SIF ratio', fontsize=16)
plt.plot(lam[beg:end], fsr, label='SCOPE Leaf SIF ratio')
plt.plot(lam[beg:end], SIFre, label='p-theory Canopy SIF ratio')
plt.plot(lam[beg:end], SIFr, label='SCOPE Canopy SIF ratio')
plt.xlabel(r'Wavelength in nm ($\lambda$)', fontsize=14)
plt.ylabel('SIF ratio', fontsize=14)
plt.legend(loc=2, fontsize=10)
plt.show()
```

```python
plt.scatter(fsr, SIFr, c='r', label='SCOPE canopy SIF ratio')
plt.scatter(fsr, SIFre, c='g', label='p-theory canopy SIF ratio')
plt.xlabel('Leaf SIF ratio', fontsize=14)
plt.ylabel('Canopy SIF ratio', fontsize=14)
plt.title('Leaf SIF ratio vs canopy SIF ratio', fontsize=16)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc=2)
plt.show()
```
