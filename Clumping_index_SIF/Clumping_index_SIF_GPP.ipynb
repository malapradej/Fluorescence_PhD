# The use of Clumping index to show how vegetation structure affects the relationship between GPP and SIF

Clumping index is a indication of the leaf spatial arrangement within the canopy, just as leaf angle distribution is an indication of the leaf normal angular arrangement within the canopy. Clumping index (CI) is a value between 0 and 1, with larger values being indicative of less clumping and smaller values increased clumping. It can be seen as a measure of leaf mutual shadowing. In canopies with more clumping (lower CI) there will be a higher canopy gap fraction and lower canopy interceptance. 

Gross Primary Productivity (GPP) is the total amount of carbon absorbed by vegetation per unit of ground area, and over a defined period. It is related to how much biomass stored within vegetation, although respiration both autotrophic and heterotrophic as well as disturbances, fires, and other processes some of which are human induces reduce the net amount of carbon absorbed by vegetation. 

Solar Induced Fluorescence (SIF) is a competing process together with heat dissipation and photosynthesis during the absorption of photosynthetically active radiation (aPAR). Under certain conditions there is a close to linear correlation between the electron transport which drives the carbon assimilation and SIF. It has been suggested that we can use the measurement of SIF to estimate the photosynthesis taking place at leaf level. 

We are attempting in this exercise to find a closer correlation between SIF and GPP at an ecosystem level than currently attempted. Several authors (Guanter et al, 2012) (Joiner et al, 2014) (Frankenberg et al, 2011) have shown that the correlation does exist using remotely sensed data of SIF and Ecosystem level models of GPP. The correlation does vary according to Plant Functional Type (PFT) with grass and crops showing a higher values of SIF and needle leaf forest lower values for the same GPP compared to all other types. We are suggesting that the difference between the types may be due to structural differences between the vegetation types more than biochemical differences. The use of CI together with SIF and GPP is to separate the amount of GPP that originates from vegetation in sun or sky light from that which is shaded within the canopy. It is known that the photosynthetic rates of shaded leaves are 2 to 3 times lower than sunlit leaves (Lichtenthaler et al, 2007) (He et al, 2012). If we find the correlation between sunlit vegetation GPP and SIF is closer than between the total GPP and SIF this should indicate that we have have corrected for at least part of the structural influence on SIF. 

Sunlit vegetation should have higher values of SIF, so we are assuming that the shaded SIF plays a lesser role in the measurements at a canopy level.


## Importing and configuring datasets

We first attempt to show the correlation between GPP and SIF per PFT. For this we need to import the GPP data for 2007 from the MPI-BGC dataset. This dataset has been shown to have the highest correlation between GPP and SIF in previous work by Guanter et al, (2012) and others. The SIF data is from the NASA GOME-2 dataset by Joiner et al. (2012). The PFT dataset is the MODIS MCD12C1 Land cover type dataset. We chose not to aggregate the PFT's into fewer grouped types as done in previous studies. We also did not destinguish between hemispheres for this study. The start for the monthly GPP and SIF dataset is the from January, and ending in December 2007. The clumping index file is for 2006. We are assuming that no sustantial changes in vegetation clumping has occured between the dates.

The different datasets have different griddings. Resampling to a common grid was undertaken prior to the running of this code. Below we read in the datasets and set constants used throughout.

```python
%matplotlib inline
import sys
import os
import numpy as np
import scipy as sp
from scipy import stats
import gdal
import osr
import ogr
import matplotlib.pylab as plt
import calendar
import pdb
plt.ion()

# datafolders

# may need to average GPP and SIF over larger areas such as in the Joiner et al (2014) paper. They used 1 deg squares.
# check the QA used on GPP....
# may need to use weekly SIF and GPP averaged over 2007 to 2011.
# checked:     check that points in all datasets are correctly co-registered.
# maybe select from weekly average SIF data and weekly GPP, but manually select areas over which to select PFT's for 
# averaging spatially.


# GPP
GPPpath = '/home/malapradej/Documents/PhD_UCL/Data/GPP/MPI-BGC/2007/'
# Laslop version of GPP. Used in Jung et al (2011) for GPP. Alternative is Reichstein version.
GPPfilename = 'GPP_GL_2007.tif' # file includes layers for Dec 2006 and Jan 2008
GPPfilename_std = 'std_GL_2007.tif' # std dev of GPP for 2007

# SIF
SIFpath = '/home/malapradej/Documents/PhD_UCL/Data/GOME-2/SIF/2007/level_3/monthly/'
SIFfilelist = 'filenames.txt' # list of files to use

# PFT
PFTpath = '/home/malapradej/Documents/PhD_UCL/Data/Land_cover/MCD12C1/2007/'
# Land cover file includes only those pixels with high QC (bit 0 and 1) and 90% pure PFT type.
PFTfilename = 'MCD12C1:Majority_Land_Cover_Type_1_2007_masked_QC_90%_Byte.tif'

# CI
CIpath = '/home/malapradej/Documents/PhD_UCL/Data/Clumping_index/'
# CI file reprojected from the MODIS sinusoidal to EPSG:4326 system and resampled to same grid as previous datasets.
CIfilename = 'MCI2_red_global_sza0_mask_reprojected_resampled.img'

# PFT area vectors
PFTvecpath = '/home/malapradej/Documents/PhD_UCL/Data/Selected_Areas/'
PFTvecfilename = 'Selected_Areas.shp'

# constants
nrows = 360 # lower resolution
nrowsBig = 3600 # higher resolution
ncols = 720 # lower resolution
ncolsBig = 7200 # higher resolution
months = 12
PFTtypes = {0:'Water', 1:'ENF', 2:'EBF' ,3:'DNF' ,4:'DBF', 5:'MXF', 6:'CSH' , 7:'OSH' , 8:'WSA' ,9:'SAV' ,10:'GRS' \
            ,11:'WET' ,12:'CRP' ,13:'URB' ,14:'CRM' ,15:'SNO' ,16:'BAR' ,17:'NAN'}
# std dev and max min limits for SIF observations
std_limit = 0.7
min_limit = 0.
max_limit = 1.5
cnt_limit = 5

# read datasets into objects

# SIF datasets in mW/m^2/nm/sr
partxt = 'NETCDF:\"%s\":SIF_740'
stdtxt = 'NETCDF:\"%s\":SIF_740_std'
cnttxt = 'NETCDF:\"%s\":Counts'
SIFfilelist = open(os.path.join(SIFpath, SIFfilelist), 'r').readlines()
SIFfilelist.sort()
SIFarr = np.zeros((nrows, ncols, months)) # SIF array in (row, col, month) format
SIFstdarr = np.zeros_like(SIFarr) # SIF std dev array
for i, f in enumerate(SIFfilelist):
    f = f.split('\n')[0] # remove newline at end
    partxt_temp = partxt % os.path.join(SIFpath, f)
    stdtxt_temp = stdtxt % os.path.join(SIFpath, f)
    cnttxt_temp = cnttxt % os.path.join(SIFpath, f)
    pararr = gdal.Open(partxt_temp).ReadAsArray()
    stdarr = gdal.Open(stdtxt_temp).ReadAsArray()
    cntarr = gdal.Open(cnttxt_temp).ReadAsArray()
    # sets bad quality values to np.nan
    pararr = np.where(pararr < min_limit, np.nan, pararr)
    pararr = np.where(pararr > max_limit, np.nan, pararr)
    pararr = np.where(cntarr < cnt_limit, np.nan, pararr)
    # sets all values above std dev limit to np.nan
    SIFarr[:,:,i] = np.where(stdarr > std_limit, np.nan, pararr)
    SIFstdarr[:,:,i] = np.where(stdarr > std_limit, np.nan, stdarr)

# Clumping index dataset between 0 and 1
CIobj = gdal.Open(os.path.join(CIpath, CIfilename))
CIarr = CIobj.ReadAsArray()/100. # convert to fraction (normalize)
CIarr = CIarr**2.

# GPP datasets in gC/m2/day
GPPobj = gdal.Open(os.path.join(GPPpath, GPPfilename))
GPPstdobj = gdal.Open(os.path.join(GPPpath, GPPfilename_std))
monthsGPP = GPPobj.RasterCount
GPParr = np.zeros((nrows, ncols, monthsGPP))
GPPstdarr = np.zeros_like(GPParr)
for i in np.arange(monthsGPP):
    temparr = GPPobj.GetRasterBand(i+1).ReadAsArray()
    GPParr[:,:,i] = np.where(temparr==-9999., np.nan, temparr) # sets all -9999 to np.nan
    temparr = GPPstdobj.GetRasterBand(i+1).ReadAsArray()
    GPPstdarr[:,:,i] = np.where(temparr==-9999., np.nan, temparr)
GPParr *= 60*60*24 * 1000. # converts units to gC/m2/day
GPPstdarr *= 60*60*24 * 1000.

# Land cover type dataset according to 18 IGBP classes
# Only pixels where QC bits 0 and 1 and > 90% of a class is present (purest pixel of type) used.
# Dataset dimensions 10 x 10 times larger than other (3600, 7200).
# Used gdal_translate and Opticks to apply band maths create masks for QC of final MOD12C1 file used.
PFTarr = gdal.Open(os.path.join(PFTpath, PFTfilename)).ReadAsArray()
```

Only those pixels of the CI, GPP, and SIF datasets where pixels overlaps completely with a pure group of PFT pixel of one type only can be selected for the training. A mask needs to be created for these pure pixels in the lower resolution products and an array for the PFT number at this lower resolution which can be used to class lower resolution pixels according to PFT. The fraction of dominant type of the pure pixel is set at 0.7 or 70%. This value was chosen so as to include as many of a single vegetation type per pixels as possible without compromising on accuracy. Although it is set at 70% there is likely to be a greater percentage of homogeneous type per pixel, due to the removel of lower resolution pixels which where not dominant (over 90% of a type) in our QC and removal stage. These pixels would include vegetation of the type of the surrounding pixel. First the pure pixel file needs to be saved to disk before vector area selection commences.

```python
frac_pure = 0.7 # fraction of pixel which should be pure, needs to be greater than 0.5
# Used a histogram binning method to select pixels based on frequency of occurance. 
type_bins = np.array(PFTtypes.keys()) # bin edges from PFT classes
type_bins = np.append(type_bins, np.max(type_bins)+1)

PFTarr_pure = np.empty((nrows, ncols), np.float16) # The purest pixel PFT array
PFTarr_pure.fill(np.nan)

# Loop over pure pixel array and higher resolution PFT arrays per histogram window
for r in np.arange(nrows):
    for c in np.arange(ncols):
        testarr = PFTarr[r*10:r*10+10,c*10:c*10+10] # histogram window in hi-res
        testarr = np.nan_to_num(testarr) # convert nan to zeros, else ignores nan's in histogram
        hist = np.histogram(testarr, type_bins, density=True)[0] # normalized histogram
        pure_type = np.where(hist >= frac_pure) # PFT class or bin with highest no pixels
        if np.shape(np.where(pure_type))[1]:
            PFTarr_pure[r, c] = pure_type[0][0]
        else:
            PFTarr_pure[r, c] = np.nan

# saves the pure pixel file to disk
driver = gdal.GetDriverByName('GTiff')
PFTfilename_pure = PFTfilename.split('.')[0]+'_%s_Pure.tif' % frac_pure
PFTfilename_pure = os.path.join(PFTpath, PFTfilename_pure)
Purepixels = driver.Create(PFTfilename_pure, ncols, nrows, 1, gdal.GDT_Int16)
transform = GPPobj.GetGeoTransform() # affine transformation parameters
Purepixels.SetGeoTransform(transform)
Purepixels.GetRasterBand(1).WriteArray(PFTarr_pure)
plt.imshow(PFTarr_pure, cmap='spectral', interpolation=None, vmin=0, vmax=17)
Purepixels = None # closes file

```

Once the pure pixel file is saved it is openend in QGIS from which the training areas are selected. Each PFT has a layer in a shapefile which delimits the pixels in the pure pixel file which will be used in the regression analysis.

```python
# Apply the selected areas vector mask for averaging
# Vector layer used to select PFT areas for averaging
attribute_filter = "Selected_A = *"
burn_value = 1
drv = gdal.GetDriverByName('MEM')
target_ds = drv.Create('', CIobj.RasterXSize, CIobj.RasterYSize, 1,  gdal.GDT_Byte)
target_ds.SetGeoTransform(CIobj.GetGeoTransform())
wgs84 = osr.SpatialReference( )
wgs84.ImportFromEPSG( 4326 )
to_proj = osr.SpatialReference()
to_proj.ImportFromWkt( CIobj.GetProjectionRef() )
target_ds.SetProjection ( to_proj.ExportToWkt() )
tx = osr.CoordinateTransformation( wgs84, to_proj )
drv = ogr.GetDriverByName( 'Memory' )
dst_ds = drv.CreateDataSource( 'out' )
dst_layer = dst_ds.CreateLayer('', srs = to_proj, geom_type=ogr.wkbPolygon ) 
vector_ds = ogr.Open( os.path.join(PFTvecpath, PFTvecfilename) )
lyr = vector_ds.GetLayer ( 0 )
lyr.SetAttributeFilter( attribute_filter )
feature = lyr.GetFeature( 0 ) 
field = feature.GetFieldDefnRef( 0 ) # make sure shapefile is closed otherwise error here
dst_layer.CreateField( field )
feature_defn = dst_layer.GetLayerDefn()
lyr.ResetReading()
for feat in lyr:
    geom = feat.GetGeometryRef()
    #geom.Transform ( tx )
    out_geom = ogr.Feature ( feature_defn )
    out_geom.SetGeometry ( geom )
    dst_layer.CreateFeature(out_geom )
    out_geom.Destroy
    geom.Destroy
dst_layer.ResetReading()
err = gdal.RasterizeLayer(target_ds, [1], dst_layer,
            burn_values=[burn_value])
if err != 0:
    print("error:", err)
PFTarr_mask = target_ds.ReadAsArray()

# Apply the selected areas mask
PFTarr_pure *= PFTarr_mask

# prints and saves the number of pixels which fit the criteria.
Pure_stats_fn = 'PFT_pure_stats_%s.txt' % frac_pure
Pure_stats_fn = os.path.join(PFTpath, Pure_stats_fn)
header = '# Number of pixels from each PFT which conform to fractional coverage.\n'
print header
file_txt = open(Pure_stats_fn, 'wt')
file_txt.write(header)
for i in PFTtypes.iterkeys():
    temp = np.shape(np.where(PFTarr_pure == i))[1]
    temp = PFTtypes[i]+'\t'+str(temp)+'\n'
    print temp
    file_txt.write(temp)
file_txt.close()
```

## Plotting and correlating SIF vs GPP per PFT

The datasets having been configured and cleaned, we now need to view the relationship between SIF per pixel and GPP for different PFT's. Monthly data between SIF and GPP is compared including the errorbars of standard deviations of GPP and SIF per month. Notice that due to the noise and requirement to average data the standard deviations are quite large. There is good agreement between the trends of the SIF and GPP data especially in the savannah, mixed forest and crop types. 

```python
# data coordinates will be saved in a dictionary with keys to PFT's and subkeys to SIF, GPP and CI.
# combined_dict format is: [PFT][SIF/GPP][pnt.nr][month.nr]
fsun = 2.5 # ratio of GPP of sun leaves compared to shaded leaves. See notes and Lichtenthaler et al (2007) 

# function for fraction of GPP which is from sunlit leaves. See notes.
hsun = lambda ci, fsun: fsun * ci / (fsun*ci + 1 - ci) 

combined_dict = {}
for i in PFTtypes.iterkeys():
    temp_dict = {}
    coords = np.where(PFTarr_pure == i)
    temp_dict['SIF'] = SIFarr[coords[0],coords[1],:] # [pnt.nr, month (12)]
    temp_dict['SIFstd'] = SIFstdarr[coords[0],coords[1],:]
    temp_dict['GPP'] = GPParr[coords[0],coords[1],:] # [pnt.nr, month (14)]
    temp_dict['GPPstd'] = GPPstdarr[coords[0],coords[1],:]
    temp_dict['CI'] = CIarr[coords[0],coords[1]] # [pnt.nr]
    h = hsun(temp_dict['CI'], fsun)
    temp_dict['GPPsun'] = np.multiply(temp_dict['GPP'].T,  h).T
    combined_dict[PFTtypes[i]] = temp_dict

# only select the PFT's below. Others are desserts etc.... also the symbols and colours for plots.
type_selected = ['ENF', 'EBF', 'DNF', 'DBF', 'MXF', 'OSH', 'WSA', 'SAV', 'GRS', 'WET', 'CRP', 'CRM']
pnt_colors = ['^g', 'vg', '^y', 'vy', '*g', 'xr', '^r', 'or', 'oy', 'oc', 'om', '*m']
lin_colors = ['-g', '--g', '-y', '--y', '-.g', '--r', '-.r', '-r', '-.y', '-c', '-m', '--m']
#type_selected = ['ENF', 'DNF', 'SAV', 'GRS']
#pnt_colors = ['^g', '^y', 'or', 'oy']
#lin_colors = ['-g', '-y', '-r', '-.y']

# average the GPP and SIF as well as std values per month
ave_dict = {}
month_list = np.arange(months)
for i in type_selected:
    pft = combined_dict[i]
    ave_sif = np.zeros(months)
    ave_sif_std = np.zeros(months)
    ave_gpp = np.zeros(months)
    ave_gpp_std = np.zeros(months)
    ave_gpp_sun = np.zeros(months)
    for j in month_list:
        ave_sif[j] = np.nanmean(pft['SIF'][:,j])
        ave_sif_std[j] = np.sqrt(np.nanmean(pft['SIFstd'][:,j]**2)) # std of a std's see:
        # http://stats.stackexchange.com/questions/25848/how-to-sum-a-standard-deviation
        ave_gpp[j] = np.nanmean(pft['GPP'][:,j])
        ave_gpp_std[j] = np.nanmean(pft['GPPstd'][:,j])
        #pdb.set_trace()
        ave_gpp_sun[j] = np.nanmean(pft['GPPsun'][:,j])
        
    ave_dict[i] = {'SIF':ave_sif,'SIFstd':ave_sif_std,'GPP':ave_gpp,'GPPstd':ave_gpp_std, 'GPPsun':ave_gpp_sun}

# plot the monthly time series of average GPP and SIF values
nrows = len(type_selected)/2
month_txt = list(calendar.month_abbr)[1:]
fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(12,12), sharex=True)
for j, i in enumerate(type_selected):
    r = j/2
    c = np.mod(j,2)
    ax1 = axes[r, c]
    ave_sif = ave_dict[i]['SIF']
    ave_sif_std = ave_dict[i]['SIFstd']
    ax1.errorbar(month_list, ave_sif, yerr=ave_sif_std, fmt='-', ecolor='r', label='SIF & err')
    #ax1.set_ylim(bottom=np.nanmin(ave_sif)-0.1, top=np.nanmax(ave_sif)+0.1)
    ax1.set_title(i)
    ax1.set_ylabel('SIF')
    ax2 = ax1.twinx()
    ave_gpp = ave_dict[i]['GPP']
    ave_gpp_std = ave_dict[i]['GPPstd']
    ave_gpp_sun = ave_dict[i]['GPPsun']
    ax2.errorbar(month_list,ave_gpp, yerr=ave_gpp_std, fmt='--', ecolor='g', label='GPP & err')
    ax2.plot(month_list,ave_gpp_sun, 'b:', label='GPP_sun')
    #ax2.set_ylim(bottom=np.nanmin(ave_gpp)-0.1, top=np.nanmax(ave_gpp)+0.1)
    ax2.set_ylabel('GPP')
    plt.xticks(month_list, month_txt)
    if i == 'CRM':
        lns = ax1
        ax1.legend(loc=1)
        ax2.legend(loc=0)
    
plt.tight_layout()
plt.show()
```

Then SIF vs GPP for the full year is compared. Linear regression is applied to the data and parameters are plotted for later comparison with clumping index corrected SIF and GPP.

```python
placing = 9.5
decrement = 0.4
fig = plt.figure(figsize=(12,12))
X = []
Y = []
slope_dict = {}
for i, pc, lc in zip(type_selected, pnt_colors, lin_colors):
    x = ave_dict[i]['SIF']
    X.append(x)
    y = ave_dict[i]['GPP']
    Y.append(y)
    mask = ~np.isnan(x) & ~np.isnan(y)
    slope, intercept, r_value, p_value, stderr = stats.linregress(x[mask], y[mask])
    slope_dict[i] = slope
    #fit = plt.polyfit(x[mask], y[mask], 1)
    fit_fn = plt.poly1d((slope, intercept))
    text = r'%s, $r^2=%.2f, y=%.2fx+%.2f$' % (i, r_value**2, slope, intercept)
    plt.text(0.04, placing, text, fontsize=14)
    placing -= decrement
    plt.plot(x, y, pc, x, fit_fn(x), lc, label=i)
X = np.array(X).flatten()
Y = np.array(Y).flatten()
mask = ~np.isnan(X) & ~np.isnan(Y)
slope, intercept, r_value, p_value, stderr = stats.linregress(X[mask], Y[mask])
slope_dict['mean'] = slope
text = r'all PFT, $r^2=%.2f, y=%.2fx+%.2f$' % (r_value**2, slope, intercept)
placing -= decrement
plt.text(0.04, placing, text, fontsize=14)
plt.tick_params(labelsize=12)
plt.xlabel('SIF', fontsize=15)
plt.ylabel('GPP', fontsize=15)
plt.ylim(top=10)
plt.legend(loc=4)
plt.show()
```

Now we need to correct for the clumping in the SIF signal. The $h_{sun}$ term was calculated previously. In fact we are trying to estimate the GPP due to direct irradiation as opposed to shaded leaves. This should correlate better with SIF. We note that the correlation coefficients ($r^2$) for individual PFT's stay the same, although the $r^2$ for all the data points combined shows an improvement, although by very little. This can only point to an improvement in finding a unique relationship between GPP and SIF through the incorporation of clumping in our estimate. We are improving the all vegetation type GPP and SIF relationship whereas individual relationships show no change as would be expected. We are obviously giving the same weight to each PFT whereas different PFT's would carry different weighting towards the overall correlation between GPP and SIF. 

```python
placing = 9.5
decrement = 0.4
fig = plt.figure(figsize=(12,12))
X = []
Y = []
slope_dict_sun = {}
for i, pc, lc in zip(type_selected, pnt_colors, lin_colors):
    x = ave_dict[i]['SIF']
    X.append(x)
    y = ave_dict[i]['GPPsun']
    Y.append(y)
    mask = ~np.isnan(x) & ~np.isnan(y)
    slope, intercept, r_value, p_value, stderr = stats.linregress(x[mask], y[mask])
    slope_dict_sun[i] = slope
    #fit = plt.polyfit(x[mask], y[mask], 1)
    fit_fn = plt.poly1d((slope, intercept))
    text = r'%s, $r^2=%.2f, y=%.2fx+%.2f$' % (i, r_value**2, slope, intercept)
    plt.text(0.04, placing, text, fontsize=14)
    placing -= decrement
    plt.plot(x, y, pc, x, fit_fn(x), lc, label=i)
X = np.array(X).flatten()
Y = np.array(Y).flatten()
mask = ~np.isnan(X) & ~np.isnan(Y)
slope, intercept, r_value, p_value, stderr = stats.linregress(X[mask], Y[mask])
slope_dict_sun['mean'] = slope
text = r'all PFT, $r^2=%.2f, y=%.2fx+%.2f$' % (r_value**2, slope, intercept)
placing -= decrement
plt.text(0.04, placing, text, fontsize=14)
plt.tick_params(labelsize=12)
plt.xlabel('SIF', fontsize=15)
plt.ylabel(r'GPP$_{sun}$', fontsize=15)
plt.ylim(top=10)
plt.legend(loc=4)
plt.show()
```
