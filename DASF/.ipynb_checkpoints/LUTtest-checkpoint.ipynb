# Testing LUTs for GOME-2

First, lets get a list of the LUTs, and make sure we can read one.

```python
import pickle
from glob import glob
%matplotlib inline
import numpy as np
import pylab as plt
import re
import pdb

# lut files containing main data per observation or two
lutFiles = glob('LUT_*.p')
# file with positional and other information needed
obsFile = 'subset_remainder.p'
# load observations dict
obs = pickle.load(open(obsFile))
# convert all toa refl to * pi
obs['Ref_toa'] = np.array(obs['Ref_toa']) * np.pi
wlmin = 680.
wlmax = 755.
```

```python
# Radiance and irradiance values
plt.figure(figsize=(10,7))
x = obs['Lam_gome'][o_nr]
y1 = obs['Irr_toa'][o_nr]
y2 = obs['Rad_toa'][o_nr]
ref = y2/y1*np.pi/np.cos(obs['Sol_zen'][o_nr]*np.pi/180.)
plt.plot(x, y1, label='Irradiance')
plt.plot(x, y2, label='Radiance')
plt.plot(x, ref, label='Reflectance')
plt.xlim(wlmin, wlmax)
plt.legend(loc='best')
plt.title('Radiance and Irradiance values', fontsize='x-large')
```

```python
# index of obs in file. see that o_nr exists in file.
br = False
for index, i in enumerate(lutFiles):
    #print i
    for j in re.findall(r'\d+', i):
        #print j
        if int(j) == o_nr:
            br = True
            break
    if br:
        break
print 'examining',lutFiles[index]
```

```python
# load the lut and remainder file into dictionaries
lut = pickle.load(open(lutFiles[index]))

# search for a subsample of the overly high resolution libRadtran data
for k in lut.keys():
    lut[k] = np.array(lut[k])
    if lut[k].ndim == 1:
        pass
    else:
        lut[k] = lut[k][:,4::9]
```

```python
# select observation indices in obs file

sza = obs['Sat_zen'][o_nr]
# select values in lut where sat za agrees with obs sat za
goodlut = np.where(lut['sat_zen'] == sza)
print goodlut
```

```python
# GIOVANNI selection
# so what water vapours and others do we have and select those which are the closest match to Giovanni data
print 'WVC'
print np.unique(lut['WVC'])
wvc = 40.
ind_wvc = np.where(np.array(lut['WVC'])== wvc)[0]
print ind_wvc
# AOTs
print 'AOT'
print np.unique(lut['AOT'])
aot = 0.2
ind_aot = np.where(np.array(lut['AOT'])== aot)[0]
print ind_aot
# pressures
print 'press'
print np.unique(lut['press'])
press = 950
ind_press = np.where(np.array(lut['press'])== press)[0]
print ind_press
# altitudes
print 'alt'
print np.unique(lut['alt'])
alt = lut['alt'][0]
ind_alt = np.where(np.array(lut['alt'])== alt)[0]
print ind_alt

# what lut index agrees with above parameters
n = list(set(ind_alt) & set(ind_press) & set(ind_aot) & set(ind_wvc))[0]
# index in lut for the GIOVANNI match
print 'GIOVANNI index in lut:', n
```

```python
# check that the lut and obs sat zen agree as rough check on same point
print 'vsen', obs['Sat_zen'][o_nr], np.unique(lut['sat_zen'][goodlut])[0]
```

```python
def surf_refl(toa_refl, atm_path, dbl_trans, spher_alb):
    '''calculate boa or surface reflectance.'''
    dr = toa_refl - atm_path
    rsurf = dr/(dbl_trans + spher_alb*dr)
    return rsurf

# now interpolate 
from scipy.interpolate import interp1d

xbaseFull = np.array(obs['Lam_gome'][o_nr])
xbase = xbaseFull[(xbaseFull>=wlmin) * (xbaseFull<=wlmax)]
ybase = obs['Ref_toa'][o_nr][(xbaseFull>=wlmin) * (xbaseFull<=wlmax)]

x0 = lut['lam'][n]
y = {}

for k in ['dbl_trans','spher_alb','atm_path']:
    y[k] = (interp1d(x0, lut[k][n], kind='cubic'))(xbase)

srefl = surf_refl(ybase, y['atm_path'], y['dbl_trans'], y['spher_alb'])

# plot all values for the GIOVANNI parameters case
plt.figure(figsize=(10,7))
plt.plot(xbase,ybase,label='toa refl')
plt.plot(xbase,y['dbl_trans']/5.,label='dbl_trans / 5.')
plt.plot(xbase,y['spher_alb'],label='spher_alb')
plt.plot(xbase,y['atm_path'] * 2,label='atm_path * 2')
plt.plot(xbase, srefl, label='GIOVANNI surf_refl')
plt.title("raz %.2f vzen %.2f szen %.2f atm %s AOT %.2f WVC %.2f"%\
          (lut['rel_azi'][n],lut['sat_zen'][n],lut['sol_zen'][n],\
           lut['atm'][n],\
           lut['AOT'][n],lut['WVC'][n]))

plt.legend(loc='best')
plt.xlim(wlmin, wlmax)
```

```python
# leaf spectrum is imported
# used the parameters provided in Knyazikhin paper refering to Lewis paper....
from scipy import optimize


leaf = np.loadtxt('leaf_spectrum.txt').T
leafLambda = leaf[0]
leafAlbedo = leaf[1] + leaf[2]

# reduce the leaf lam and albedo data size (speeds up interpolation...)
xleafFull = leafLambda
leafLambda = xleafFull[(xleafFull>=wlmin) * (xleafFull<=wlmax)]
leafAlbedo = leafAlbedo[(xleafFull>=wlmin) * (xleafFull<=wlmax)]

# function for Lewis 8 parameter model
def lewis_8(a, b, c, a_, s, b_, c_, d_):
    '''the 8 parameter Lewis function for canopy reflection. see Lewis et al (2003)'''
    refl = a * leafW + b * leafW**2 / (1 - c*leafW) + a_*s + leafW*s * (b_ + d_*s) / (1 - c_*leafW)
    return refl

# cost function for first minimization
# the first minimization is for atmospheric parameter selection
# the second would be for the selection of a and p....
def mismatch_function(p, leafW, y):
    '''equation (1) in Lewis et al (2003).
    This is a spectral invariant solution including surface reflectance.
    '''
    a = p[0] # canopy first order scattering
    b = p[1] # canopy second order scattering
    c = p[2] # 
    a_ = p[3]
    s = p[4]
    b_ = p[5]
    c_ = p[6]
    d_ = p[7]
    y_hat = lewis_8(a, b, c, a_, s, b_, c_, d_)
    #y_hat = p[0] * leafW / (1 - p[1] * leafW) + p[2]
    diff = (y_hat - y)
    return diff
```

```python
# goodlut is the lut points associated with this sample
# sample  is the observation sample number

# the sum of squared errors for 1st minimization
def sse(p,leafW,y):
    '''Sum of squared error'''
    return (mismatch_function(p,leafW, y)**2.).sum()

# there are 8 parameters in the Lewis model. a to c and a_ to d_ and soil refl.
bound = np.array([(0.001,0.999)]*8)

results = []
results2 = []

# leaf albedo interpolated to obs lam
leafW = (interp1d(leafLambda, leafAlbedo, kind='cubic'))(xbase)

def mismatch_function2(p, leafW, y):
    '''the 2nd cost function for the retrieval of a and p.'''
    a = p[0] # canopy interceptance * directional gap density
    p = p[1] # recollision probability
    y_hat = a * leafW / (1 - p*leafW)
    #y_hat = p[0] * leafW / (1 - p[1] * leafW) + p[2]
    diff = (y_hat - y)
    return diff
    
def sse2(p2,leafW,y):
    '''Sum of squared error for the 2nd cost function'''
    return (mismatch_function2(p2,leafW, y)**2.).sum()
```

```python
# loop over all values in lut for this point and minimize for each the values of a and p
# the best fit for the parameters (a and p) against observations will be selected per RT parameters
for i in goodlut[0]: # goodlut is tuple need to select 1st element
    #plt.figure(figsize=(10,7))
    #print '========',i,'========'
    y = {}
    x0 = lut['lam'][i]
    
    for k in ['dbl_trans','spher_alb','atm_path']:
        y[k] = (interp1d(x0, lut[k][i], kind='cubic'))(xbase)
    
    rsurf = surf_refl(ybase, y['atm_path'], y['dbl_trans'], y['spher_alb'])
    
    p = [0.5]*8
    
    psolve = optimize.fmin_l_bfgs_b(sse,p,approx_grad=True,iprint=-1,\
                                args=(leafW,rsurf),bounds=bound)  
    print i,'sse:', psolve[1], 'a,b,c,a_,s,b_,c_,d_:', psolve[0]
    
    fwd_c = psolve[0][0] * leafW + psolve[0][1] * leafW**2 / (1 - psolve[0][2]*leafW)
    
    p2 = [0.5]*2
    
    bound2 = np.array([(0.001,0.999)]*2)
    
    psolve2 = optimize.fmin_l_bfgs_b(sse2,p2,approx_grad=True,iprint=-1,\
                                args=(leafW,fwd_c),bounds=bound2)  
    print i,'sse2:', psolve2[1], 'a,p:', psolve2[0]

    
    results.append(psolve)
    results2.append(psolve2)
    
```

```python
# select parameter set that gives minimum rmse
rses = np.zeros((len(results), 2))
for i, (j, k) in enumerate(zip(results, goodlut[0])):
    rses[i, 0] = j[1]
    rses[i, 1] = k
ind_min = np.argmin(rses[:,0])
print 'nr:', rses[ind_min, 1]
print 'a,b,c,a_,s,b_,c_,d_:', results[ind_min][0]
print 'sse:', results[ind_min][1]
print 'a,p:', results2[ind_min][0]
print 'sse2:', results2[ind_min][1]
print 'wvc:', lut['WVC'][ind_min]
print 'press:', lut['press'][ind_min]
print 'aot:', lut['AOT'][ind_min]
print 'alt:', lut['alt'][ind_min]
print 'clf:', obs['Clf'][o_nr]
print 'sat zen:', obs['Sat_zen'][o_nr]
print 'lat:', obs['Lat'][o_nr]
print 'lon:', obs['Lon'][o_nr]
```

```python
min_solve = results[ind_min]
sse_min = min_solve[1]
param = min_solve[0]
a = param[0] 
b = param[1]
c = param[2]
a_ = param[3]
s = param[4]
b_ = param[5]
c_ = param[6]
d_ = param[7]

fwd = lewis_8(a, b, c, a_, s, b_, c_, d_)
#fwd = param[0] * leafW / ( 1 - param[1] * leafW) + param[2]

# only the canopy contribution
fwd_c = a * leafW + b * leafW**2 / (1 - c*leafW)

# 2nd estimate of parameters
a = results2[ind_min][0][0]
p = results2[ind_min][0][1]
fwd_c2 = a * leafW / (1 - leafW * p)

DASF = a / (1 - p)
print 'DASF:', DASF

#param_test = [0.093, 0.663, 0.1]
#fwd_test = param_test[0] * leafW / ( 1 - param_test[1] * leafW) + param_test[2]

# surf refl
for k in ['dbl_trans','spher_alb','atm_path']:
    y[k] = (interp1d(x0, lut[k][ind_min], kind='cubic'))(xbase)
  
dr = ybase - y['atm_path']
rsurf = dr/(y['dbl_trans'] + y['spher_alb']*dr)

plt.figure(figsize=(10,7))
plt.plot(xbase,ybase, label='toa refl')
plt.plot(xbase,rsurf, label='surf refl')
plt.plot(xbase,fwd, label='8 para surf refl')
#plt.plot(xbase, fwd_test, '--r')
plt.plot((wlmin, wlmax), (a_*s, a_*s), label='soil refl')
plt.plot(xbase, fwd_c, '--r', label='8 para canopy refl')
plt.plot(xbase, fwd_c2, '--b', label='2 para canopy refl')
plt.plot(xbase, leafW*0.1)
plt.legend(loc='best')

# LAI from p 
L = (np.log(1 - p/0.88)/-0.7)**(4./3.)
print 'LAI:', L

plt.xlim(wlmin, wlmax)
print 'sse:', sse_min
```
