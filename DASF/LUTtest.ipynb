```python
# the slit function 

def slit_smooth(lam, spectrum, fwhm, width):
    '''
    Function that applies the gaussian slit function with known FWHM to spectrum.
    It will replace the boundaries of the smoothed spectrum which are affected by
    zero padding with the original spectrum.
    Input:
    lam - wavelenght array in nm
    spectrum - the spectrum to convolve
    fwhm - the full width half maximum of the gaussian in nm
    width - the width of the slit function in nm
    Output:
    convol - smoothed spectrum
    '''
    resol = np.average(np.diff(lam)) # average spectral resol of model in nm
    intervals = np.ceil((width / resol)+1) # the number of intervals in the width of slit
    slit = signal.gaussian(intervals, fwhm/2.3548201*intervals/width)
    convol = signal.fftconvolve(spectrum, slit/np.sum(slit), mode='same')
    hwidth = (intervals-1)/2
    convol[:intervals] = spectrum[:intervals]
    convol[-intervals-1:] = spectrum[-intervals-1:]
    return convol
```

```python
# surface reflectance including cloud correction model PC1 only

def p_srefl(a, p, leafW):
    '''p-theory based surface reflectance.'''
    sref = a * leafW / (1 - p*leafW)
    return sref

def toa_refl_pca(a, p, clf, b, pc, atm_path, dbl_trans, spher_alb, leafW):
    '''
    toa reflectance using pc1 cloud correction.
    See notes on 30/05/15.
    '''
    sref = p_srefl(a, p, leafW)
    rtoa = (1 - clf)*(atm_path + dbl_trans*sref/(1 - spher_alb*sref)) + clf*(b*pc)
    return rtoa

# load the pca data file into a Dataframe
h5_fn = 'cloud_sims.h5'
pca_df = pd.read_hdf(h5_fn, 'pca')

# read pc's into arrays
lam_pc = pca_df.index
pcs = pca_df.values.T

# interpolate pc's and wavelength to observations
pcs = (interp1d(lam_pc, pcs, kind='linear'))(xbase)

# apply slit function to modelled pcs
for i in np.arange(len(pcs)):
    pcs[i] = slit_smooth(xbase, pcs[i], fwhm, width)

def sse_pca(pars, pc, atm_path, dbl_trans, spher_alb, leafW, ref_obs):
    '''Sum of squared error cost function for cloud included solution'''
    a = pars[0]
    p = pars[1]
    clf = pars[2]
    b = pars[3]
    return (mismatch_function(a, p, clf, b, pc, atm_path, dbl_trans, spher_alb, leafW, ref_obs)**2.).sum()

# bounds of the a, p, clf and b parameters
bound = np.array([(0.001,0.999),(0.001,0.999),(0.000,1.000),(0.0,110.)])

results = []

# leaf albedo interpolated to obs lam
leafW = (interp1d(leafLambda, leafAlbedo, kind='linear'))(xbase)

def mismatch_function(a, p, clf, b, pc, atm_path, dbl_trans, spher_alb, leafW, ref_obs):
    '''cost function for the retrieval of a, p, clf and b'''
    ref = toa_refl_pca(a, p, clf, b, pc, atm_path, dbl_trans, spher_alb, leafW)
    diff = np.abs(ref - ref_obs)
    return diff

clf_gome = obs['Clf'][o_nr] # cloud fraction read from GOME-2 data

# goodlut is the lut points associated with this observation
# loop over all values in lut for this point and minimize for each the values of a and p
# the best fit for the parameters (a and p) against observations will be selected per RT parameters
for i in goodlut[0]: # goodlut is tuple need to select 1st element
    y = {}
    x0 = lut['lam'][i] # the full wavelength range of modelled lut data
    
    # interpolate modelled data to observation wavelengths and set to observation limits
    # apply slit function to modelled values    
    for k in ['dbl_trans','spher_alb','atm_path']:
        y[k] = (interp1d(x0, lut[k][i], kind='linear'))(xbase)
        y[k] = slit_smooth(xbase, y[k], fwhm, width)
    
    # prior estimates of parameters
    # [a, p, clf, b]
    pars = [0.1, 0.8, clf_gome, 65.0]

    # bfgs is very sensitive to the initial guess. results differ a lot between changes in prior estimates of parameters
    # not very sensitive to changes in b. advantage is that b can be set to an approximate cloud reflectance. faster.
    
    psolve = optimize.fmin_l_bfgs_b(sse_pca,pars,approx_grad=True,iprint=-1,\
                                args=(pcs[0], y['atm_path'], y['dbl_trans'], y['spher_alb'],\
                                      leafW, ybase),bounds=bound)  
    
    # fmin (Nelder-Mead Simplex) seems to give clf values closer to GOME-2. slower. more sensitive to changes in b.
    # changes in prior estimates does not affect the final solution as much as bfgs.
    '''
    psolve = optimize.fmin(sse_pca,pars,args=(pcs[0], y['atm_path'], y['dbl_trans'], y['spher_alb'],\
                                      leafW, ybase),full_output=True, disp=False)
    '''
    # slowest. clf retrieved not as good as fmin. solution of a, p and b similar to fmin.
    '''
    psolve = optimize.fmin_powell(sse_pca,pars,args=(pcs[0], y['atm_path'], y['dbl_trans'], y['spher_alb'],\
                                      leafW, ybase),full_output=True, disp=False)
    '''
    
    # print out the solutions per lut entry used in selecting the minimum rmse. uncomment to use
    '''
    print i,'sse:', psolve[1], 'a,p,clf,b:', psolve[0]
    
    results.append(psolve)
    '''
    results.append(psolve)
    
# select parameter set that gives minimum rmse and print out
rses = np.zeros((len(results), 2))
for i, (j, k) in enumerate(zip(results, goodlut[0])):
    rses[i, 0] = j[1]
    rses[i, 1] = k
ind_min = np.argmin(rses[:,0])
nr = rses[ind_min, 1]
print 'nr:', nr
a = results[ind_min][0][0]
print 'a: %.4f' % a
p = results[ind_min][0][1]
print 'p: %.4f' % p
clf = results[ind_min][0][2]
print 'clf: %.4f' % clf
b = results[ind_min][0][3]
print 'b: %.4f' % b
sse = results[ind_min][1]
print 'sse:', sse
wvc = lut['WVC'][ind_min]
print 'wvc:', wvc
press = lut['press'][ind_min]
print 'press:', press
aot = lut['AOT'][ind_min]
print 'aot:', aot
alt = lut['alt'][ind_min]
print 'alt:', alt
print 'clf_gome:', clf_gome
sat_zen = obs['Sat_zen'][o_nr]
print 'sat zen:', sat_zen
lat = obs['Lat'][o_nr]
print 'lat:', lat
lon = obs['Lon'][o_nr]
print 'lon:', lon

min_solve = results[ind_min]
sse_min = sse
param = min_solve[0]

# forward model of p-theory surf refl
fwd = p_srefl(a, p, leafW)
#fwd = a * leafW / (1 - leafW * p)

# DASF...
DASF = a / (1 - p)
print 'DASF:', DASF

# surf refl and
# apply slit function to modelled values
for k in ['dbl_trans','spher_alb','atm_path']:
    y[k] = (interp1d(x0, lut[k][ind_min], kind='linear'))(xbase)
    y[k] = slit_smooth(xbase, y[k], fwhm, width)
    
def surf_refl_pca(clf, b, pc, atm_path, dbl_trans, spher_alb, toa_refl):
    '''
    surface reflectance using pc1 cloud correction.
    See notes on 30/05/15.
    '''
    B = (toa_refl - clf*b*pc - atm_path*(1 - clf))/((1 - clf)*dbl_trans)
    rsurf = B / (1 + spher_alb*B)
    return rsurf

# forward model surface reflectance using minimized parameters
rsurf = surf_refl_pca(clf, b, pcs[0], y['atm_path'], y['dbl_trans'], y['spher_alb'], ybase)

# calculate p curve at wider range
'''
leaf = np.loadtxt('leaf_spectrum.txt').T
leafLambda = leaf[0]
leafAlbedo = leaf[1] + leaf[2]
pindex = (leafLambda >= 600) * (leafLambda <= 800)
px = leafLambda[pindex]
pleaf = leafAlbedo[pindex]
py = a * pleaf / (1 - pleaf * p)
'''

plt.figure(figsize=(10,7))
plt.plot(xbase,ybase, label='toa refl')
plt.plot(xbase,rsurf, label='surf refl')
plt.plot(xbase, fwd, 'r-', label='4 para canopy refl')
plt.plot(xbase, leafW*0.2, '--g', label='leaf albedo/5')
plt.plot(xbase, clf*b*pcs[0], '--b', label='cloud refl')
plt.plot(wls, mode, 'r--', label='LER')
plt.legend(loc='best')

#plt.plot(px, py, 'r--', label='2 para wider')

# LAI from p 
def LAI_p(p):
    '''LAI from recollision prob.
    see Smolander & Stenberg (2005)'''
    L = (np.log(1 - p/0.88)/-0.7)**(4./3.)
    return L

print 'LAI:', LAI_p(p)

plt.xlim(wlmin, wlmax)

# append cloud fraction data to text file for later correlation
'''
clf_fn = 'cloud_fractions_pc1.cvs'
if os.path.exists(clf_fn):
    os.remove(clf_fn)
text = '%d, %.5f, %.5f\n' % (o_nr, clf, clf_gome)
clf_fl = open(clf_fn, 'a')
clf_fl.write(text)
clf_fl.flush()
os.fsync(clf_fl.fileno())
clf_fl.close()
'''
```
